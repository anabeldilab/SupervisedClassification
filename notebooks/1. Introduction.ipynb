## Descripción del Conjunto de Datos

El conjunto de datos utilizado es el de [Chest X-Ray Pneuminia](https://www.kaggle.com/datasets/paultimothymooney/chest-xray-pneumonia) disponible en Kaggle, compuesto por 5863 imágenes de radiografías. 

El conjunto de datos está estructurado en dos carpetas principales:
- Una para **imágenes de entrenamiento**, que será utilizada para entrenar nuestros modelos de clasificación.
- Otra para **imágenes de prueba**, que será utilizada para evaluar el rendimiento de los modelos.

Dentro de cada una de estas carpetas, las imagenes se encuentran divididas en dos subcarpetas, coincidiendo con las dos clases a clasificar:
- **Pneumonia**: Imágenes de pacientes con neumonía.

- **Normal**: Imágenes de pacientes sin neumonía.
